```python
import pandas as pd
import numpy as np
data = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
def sniff_modified(df):
    with pd.option_context("display.max_colwidth", 20):
        info = pd.DataFrame()
        info['data type'] = df.dtypes
        info['percent missing'] = df.isnull().sum()*100/len(df)
        info['No. unique'] = df.apply(lambda x: len(x.unique()))
        info['unique values'] = df.apply(lambda x: x.unique())
        return info.sort_values('data type')
data
sniff_modified(test_df)
```

```python
from pandas.api.types import is_numeric_dtype, is_string_dtype, is_object_dtype
def fix_missing_num(df, colname):
    df[colname].fillna(df[colname].median(), inplace=True)
    return df[colname].median()
import category_encoders as ce
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace = True)
X_valid = test_df
```

```python
X = data.drop('Survived', axis = 1)
y = data['Survived']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.01, random_state =12, stratify = y)
X_train.Sex = X_train.Sex.map({'male': 1, 'female': 0})
X_train['Age_Fill'] = X_train['Age'].groupby([X_train['Pclass'], X_train['Sex']]).apply(lambda x: x.fillna(x.median()))
X_test.Sex = X_test.Sex.map({'male': 1, 'female': 0})
X_test['Age_Fill'] = X_train['Age'].groupby([X_train['Pclass'], X_train['Sex']]).apply(lambda x: x.fillna(x.median()))
X_valid.Sex = X_valid.Sex.map({'male': 1, 'female': 0})
X_valid['Age_Fill'] = X_train['Age'].groupby([X_train['Pclass'], X_train['Sex']]).apply(lambda x: x.fillna(x.median()))
X_train['CabinClass'] = X_train['Cabin'].str[0:1]
X_train = X_train.fillna({'CabinClass' : 0})
X_test['CabinClass'] = X_test['Cabin'].str[0:1]
X_test = X_test.fillna({'CabinClass' : 0})
X_valid['CabinClass'] = X_valid['Cabin'].str[0:1]
X_valid = X_valid.fillna({'CabinClass' : 0})
```

```python
for colname in X_train.columns:
    if is_numeric_dtype(X_train[colname]):
        med = fix_missing_num(X_train, colname)
        print (med)
        X_test[colname].fillna(med, inplace = True)
        X_valid[colname].fillna(med, inplace = True)
X_train.info()
```

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, BatchNormalization, Dropout
from tensorflow.keras.callbacks import EarlyStopping,ModelCheckpoint
stop = EarlyStopping(monitor='val_loss', mode = min, patience = 500)
model = Sequential()
model.add(Dense(256, input_dim = 20, activation = 'relu'))
model.add(Dense(256, activation = 'softmax'))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(64, activation = 'softmax'))
model.add(Dense(64, activation = 'softmax'))
model.add(Dense(1, activation = 'sigmoid'))
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
mc = ModelCheckpoint('best_model.h5', monitor='val_accuracy', mode='max', verbose=1, save_best_only=True)
model.summary()
```

```python
history = model.fit(X_train, y_train, epochs = 4000, validation_split = 0.1, callbacks = [stop, mc])
```

```python
_, train_acc = model.evaluate(X_train, y_train, verbose=0)
_, test_acc = model.evaluate(X_test, y_test, verbose=0)
print('Train: %.3f, Test: %.3f' % (train_acc, test_acc))
```
